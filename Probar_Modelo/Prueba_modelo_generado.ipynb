Clasificador de los tipos de Pokemon de Agua y Fuego, para probar el modelo generado.

Importamos las librerias que vamos a utilizar.

```python
from keras.preprocessing import image 
from keras.utils import load_img
from keras.utils import img_to_array
from keras import models
import numpy as np
import matplotlib.pyplot as plt
```

Importamos el modelo que creamos

```python
model = models.load_model("./modelo.h5")
```

Primero importamos nuestra imagen que queremos probar. Estas imagenes fueron descargadas desde internet y de varios tipos para poder realizar las pruebas. Después preprocesamos la imagen para indicarle un tamaño, para posteriormente indicarle la respuesta del modelo que importamos y al ser un clasificador binario le indicamos que al ser menor a 0.5 que salga un resultado de tipo Fire del Pokémon y si es mayot a 0.5 nos saldra una predicción de Pokémon tipo Water. Por último, ploteamos la imágen y el resultado de la predicción.

```python
test_img = './images/charizard_3d.png'
# test_img = './images/charizard_funko.jpg'
# test_img = './images/charizard_video.png'
# test_img = './images/vaporeon_botarga.jpg'
# test_img = './images/vaporeon_furro.png'
# test_img = './images/vaporeon_guapo.png'
# test_img = './images/vaporeon_peluche.png'
# test_img = './images/vaporeon_tarjeta.png'
# test_img = './images/lucario.png'
# test_img = './images/lucario_mega.png'
# test_img = './images/fuecoco.png'
# test_img = './images/quaxly.png'
# test_img = './images/gato.png'
# test_img = './images/camaron.png'

img_path = test_img 

img = load_img(img_path,  target_size=(256,256))
img_tensor = img_to_array(img)
img_tensor = np.expand_dims(img_tensor, axis = 0)
img_tensor /= 255.

confidence = model.predict(img_tensor)
predict_class = (confidence > 0.5).astype("int32")
print (confidence)
print ("class ", predict_class[0][0], "confindence", )

if predict_class[0][0] == 0:
  print("Fire")
else:
  print("Water")

plt.imshow(img_tensor[0])
plt.show()
```
